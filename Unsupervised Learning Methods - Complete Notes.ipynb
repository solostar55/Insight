## Silhouette Coefficient

* You enter the X-array, and the predicted clusters (model.labels_)
* also, enter the distance metric ==> "euclidean"

## DOING A LOOP TO CALCULATE THE N_CLUSTERS WITH THE BEST SILHOUETTE



```python
nclusters = np.arange(2,7)

silhouettes_complete = []
silhouettes_single = []
silhouettes_ward = []

for i in nclusters:
    
    agglom_model1 = AgglomerativeClustering(n_clusters = i, linkage="complete", affinity="euclidean")
    agglom_model2 = AgglomerativeClustering(n_clusters = i, linkage="single", affinity="euclidean")
    agglom_model3 = AgglomerativeClustering(n_clusters = i, linkage="ward", affinity="euclidean")
    
    agglom_model1.fit(xtrain)
    agglom_model2.fit(xtrain)
    agglom_model3.fit(xtrain)
    
    silhouette_complete = silhouette_score(xtrain, agglom_model1.labels_, metric="euclidean")
    silhouette_single = silhouette_score(xtrain, agglom_model2.labels_, metric="euclidean")
    silhouette_ward = silhouette_score(xtrain, agglom_model3.labels_, metric="euclidean")
    
    silhouettes_complete.append(silhouette_complete)
    silhouettes_single.append(silhouette_single)
    silhouettes_ward.append(silhouette_ward)
    
full_results = pd.DataFrame({"complete":silhouettes_complete,"single":silhouettes_single,"ward":silhouettes_ward})

full_results.plot()
```

## CLUSTER LABELS IN HIERARCHICAL CLUSTERING

- We can extract clusters at intermediate stages of a hierarchical clusteringn
- We choose a HEIGHT on the dendogram
- We see which observations are clustered at the HEIGHT
    - What is the meaning of the HEIGHT - the y axis on a dendogram?
        - it is the distance between merging clusters
        - clusters joined at lower heights are closer to each other
        - in the grain example - at the height of 5 units apart, most clusters were joined together
        - so, these clusters were joined together at a relatively low/small height - meaning they are quite similar
- HEIGHT corresponds to the DISTANCE between clusters
    - the distance between clusters is defined by a linkage method
    - in complete linkage - distance between clusters is the MAXIMUM distance between their samples (the samples within the clusters)
    - there are other linkage methods and they can give different results


### Which clusters are closest?
* The inkage method defines how the distance between clusters is measured. In complete linkage, the distance between clusters is the distance between the furthest points of the clusters. 
* In single linkage, the distance between clusters is the distance between the closest points of the clusters.
* Therefore, in complete linkage: clusters are joined together based the SMALLEST distance between the FARTHEST points inside each cluster >> essentially - closer clusters are joined. Complete linkage is just a method of defining what is "close". 


### What are the cluster labels at intermediate stages??

```python
# USING THE GRAINS DATA

grains = pd.read_csv("Datasets/Course 16 Unsupervised Learning/seeds.csv", header=None)
samples = grains

# create a separate dataframe that holds just the labels

labels = samples.iloc[:,7].replace({1:"kama",2:"rosa",3:"canadian"})

# convert that to an array
labels = labels.values

```

```python
## NOW WE CAN GET THE CLUSTER LABELS

from scipy.cluster.hierarchy import fcluster

# Extract the cluster labels, based on the height of 6
# So we stop the algorithm at a height of 6

mergings = linkage(samples, method='complete')
clusters = fcluster(mergings, 8, criterion='distance')

# clustering still occurs based on similarity. However, the distance between clusters used to determine
# whether clusters should be merged is demermined by the method 

print(clusters) # we can see below at the HEIGHT of 8, THREE types of clusters have been defined!

# Now compare to the actual labels

pred_actual = pd.DataFrame({"actual":labels,"predicted":clusters})

pd.crosstab(pred_actual["actual"],pred_actual["predicted"])

```

# Building Recommender Systems using NMF

- Finding similarities between articles
- So we can recommend similar articles to readers of other articles

### how?
- We apply NMF to the word-frequency-array
- Remember the NMF feature values describe the topics - i.e., each NMF component is a DIFFERENT TOPIC, and words (along the columns) align with those topics
    - Remember there are two sets of output
        - When we transform the data, we get the loadings of each article/document (ROWS) onto the components (COLUMNS). Therefore, we can see whether there are documents/articles that share similar patterns of loadings on the components. We can also interpret the components by seeing which articles load onto them as we did above
        - We also get the model.components_ output; this shows you how each individual word loads onto each of the components. ROWS are components, COLUMNS are words. 
        
    
- Therefore, as stated above, similar articles, will have similar NMF components
    - similar articles/documents will load onto similar components and combinations of components
- How do we compare the NMF components between two articles to determine they are similar??




```python
# below is duplicating the NMF of the articles data

from sklearn.decomposition import NMF 

# 1. Create the NMF object - extracting SIX components

model = NMF(n_components=6)  

# 2. Fit the model to the data to derive the 6 components

model.fit(articles)

# 3. Transform the data to derive factor scores of each document - documents are rows, components/topics are columns

nmf_features = model.transform(articles)

print(nmf_features.shape)
```

### DERIVING SIMILARITY
* Now we need to compare the articles in this array
* Similar articles in this array will have similar patterns of scores along each of the components

* two similar articles - can refer to similar topics (the columns/components) but refer to these topics in different strengths (i.e., have different loadings on those topics/components)

### DISTANCE
* To compare articles in this array we use the cosine similarity - this is examining the angle between the lines
* Each line is a document
* Smaller angle between two lines (or documents) yields a larger cosine value, and thus a greater similarity between those documents

```python
#calculating the cosine similarities

from sklearn.preprocessing import normalize

# 1. first normalize all the features

norm_features = normalize(nmf_features)

# 2. Recreate the data frame earlier - article scores on each of the components is the data
# and the INDEX is the title of the documents

topics_titles = pd.DataFrame(norm_features, index=titles) 

# 3. Now index a specific title

current_article=topics_titles.loc['Cristiano Ronaldo',:]

#calculate cosine similarity of this article to ALL the others

similarities = topics_titles.dot(current_article) 

print(similarities) # what is printed is the cosine similarity of the "current_article" to each of the others

# we can use nlargest to get the article articles with the largest cosine similarity to this Ronaldo article

top5 = similarities.nlargest(5)
pd.DataFrame(top5)
```

```python
# let's do the above for several of the topics - i.e., the first 5 topics!!

results = []

for i in range(0,6):
    similarities = topics_titles.dot(topics_titles.iloc[i,:])
    top_similarities = similarities.nlargest()
    results.append(top_similarities)

results
```

```python
artists = pd.read_csv("Datasets/Course 16 Unsupervised Learning/music.csv")

# we need to reshape this dataset so that the rows are the artists, the columns are the users
# and the values the number of times each user listened to that artist

artists.head()
```

```python
artists_pivoted = artists.pivot_table(index = "artist_offset",
                                     columns = "user_offset",
                                     values = "playcount",
                                     aggfunc="count")

artists_matrix = artists_pivoted.reset_index(drop=True)
```

```python
artists_matrix.head() # we now have a sparse matrix!
```

```python
artists_matrix.fillna(0, inplace=True)
```

```python
from sklearn.decomposition import NMF
from sklearn.preprocessing import Normalizer, MaxAbsScaler
from sklearn.pipeline import make_pipeline

scaler = MaxAbsScaler() # rescale the matrix before factor analysis

nmf = NMF(n_components = 20) # initialise the NMF object to extract 20 components

normalizer = Normalizer() # Normalize the component scores (the loadings of each artist on each component/topic/genre)

pipeline = make_pipeline(scaler, nmf, normalizer)

# Now we fit the data

components = pipeline.fit_transform(artists_matrix)
```

```python
# view the results

print(components.shape) # 111 rows for the 11 artists - loading onto the 20 components

# let's turn these results into a data frame

# I have to load in the artists names

artists = pd.read_csv("Datasets/Course 16 Unsupervised Learning/artists.csv", header=None)
artists = artists.values

# Now make the data frame

artists_and_components = pd.DataFrame(components, index = artists)

# view the results

artists_and_components
```
