### Build a data frame from a list of lists

```python
#buidling a data frame from lists

list1=[1,3,5] #EACH LIST WILL BECOME A COLUMN IN THE DATA FRAME
list2=[3,4,5]


list_labels=["col1","col2"] #will be the column names

list_cols=[list1,list2]

#now use zipped to create tuples

zipped=list(zip(list_labels,list_cols)) #each column name is paired with its column list as a tuple
#we are zipping up a list of KEYS, and a list of VALUES (as lists)

#convert the zipped item into a dictionary

data=dict(zipped) #converts to a dictionary

print("zipped",zipped)
print("data",data)

users=pd.DataFrame(data) #converts the dictionary to a data frame

print("\n dataframe \n\n",users)
```

### practice converting list of lists to a data frame (via ZIPPED and DICT)

```python
column_names = ["animals","preference"]

animals = ["dogs","cats","horses"]

preference = ["highest","lowest","middle"]

columndata = [animals,preference]

#zipped column names and columndata

zippeddata = list(zip(column_names,columndata))

print(zippeddata)

#convert to dictionary

dictionary = dict(zippeddata)

print(dictionary)

dataframe_animals = pd.DataFrame(dictionary)

dataframe_animals
```

```python
# More practice converting lists of lists to data frames

colnames = ["food","liking"]
foods = ["pasta","rice","pizza"]
liking = [8,9,10]

# 1. make list of lists first

dataset = [foods,liking]

# 2. zip and list the list of lists with the colnames

zippeddataset = list(zip(colnames, dataset))

# 3. convert to dictionary 

zippeddictionary = dict(zippeddataset)

# 4. Convert to data frame

pd.DataFrame(zippeddictionary)
```

# REPLACING VALUES - using .replace()
* requires you to input a dictionary

```python
def categories(dataframe,oldcolumn):
    
    speciesname = []
    
    for row in dataframe[oldcolumn]:
    
        if row == 0:
            
            newvalue = "setosa"
            
        elif row == 1:
            
            newvalue = "versicolor"
            
        elif row == 2:
            
            newvalue = "virginica"
    
        speciesname.append(newvalue)
        
    return speciesname

        
iris_df["species_name"] = categories(iris_df,"species")


iris_df.head()
```

```python
results = [] # IF I PUT THIS LIST INSIDE THE FOR LOOP, I DONT HAVE TO REPEAT THE RESULTS.APPEND command in the for loop
for row in iris_df["species"]:
    if row == 0:
        value = "setosa"
        results.append(value)
    elif row == 1:
        value = "versicolor"
        results.append(value)
    elif row == 2:
        value = "virginica"
        results.append(value)

results
        
#iris_df["species_name"] = categories(iris_df,"species")


#iris_df.head()
```

### Use of the ROLLING AVERAGE

- this is similar to downsampling, instead all rows are printed, only the average at 7 days (or whatever value for the window) is retained in the dataset
- all the previous rows now have missing data


```python
# The weather dataset is currently HOURLY
weatherdf.head()
```

```python
# We will compare the daily means using a nonsmoothed and smoothed (rolling average) aggregation

# extract two week's worth of data

unsmooth = weatherdf.loc["August 1 2010":"August 15 2010","Temperature"]

smooth = unsmooth.rolling(window=24).mean() #the rolling average smoothes out the hourly averages across the day

print(unsmooth.head())
print(smooth.head(30)) # as shown below, the smooth dataset smoothes over values in a 24 hour window, beginning AT
# the 24th hour - the result is that there are less hourly fluctuations compared to smooth

# Below I obtain a third series, which is the DAILY MEAN (so no hourly values)

unsmoothmean = unsmooth.resample("D").mean()
print(unsmoothmean.head())
```

```python

## Create a data frame combining the unsmooth and smooth - both are series, same index, so use the dictionary method

combined = pd.DataFrame({"smooth":smooth,"unsmooth":unsmooth,"unsmoothmean":unsmoothmean})
```

```python
# now plot

import pandas as pd

import matplotlib.pyplot as plt

combined.plot()
plt.show()

#the smoothed data frame plots a smooth version of the daily values!
```

```python
smooth.plot() #lets look at the smooth plot only that used the rolling average
# because it averages across 24 hours, the fluctations are less striking than the hourly averages
```

```python
unsmoothmean.plot(color="green")
```

```python
# we can also apply rolling averages to not only the means, but also the maximum

august = weatherdf.loc["August 2010","Temperature"] #extract august only


# get the daily maximums in August

dailyaughigh = august.resample("D").max()

print(dailyaughigh.head())
```

### APPLY the ROLLING AVERAGE TO THE DAILY MAXIMUMS
* in this example we obtain the daily maximum values first (which involves resampling)
* we then obtain the rolling mean of these daily maximums over a 7-day period (as opposed to a 30 day period whcih was used above).

```python
# Below - get the daily maximums (data is initially hourly)
# Convert these daily maximums into a weekly rolling average which smoothes out the daily averages

dailyaughigh_smooth = august.resample("D").max().rolling(window=7).mean() 

print(dailyaughigh_smooth.head(14)) #See below - the first rolling mean begins at 7 days
# values in between are all smoothed (these are no longer the RAW daily maximums)

dailyaughigh_smooth.plot() # we can plot the weekly rolling average
```

## MORE COMPLEX EXAMPLES

BASIC SYNTAX

.             One character except new line
\.            A period. \ escapes a special character.
\d            One digit
\D            One non-digit
\w            One word character including digits
\W            One non-word character
\s            One whitespace
\S            One non-whitespace
\b            Word boundary
\n            Newline
\t            Tab

MODIFIERS

$             End of string
^             Start of string
ab|cd         Matches ab or de.
[ab-d]	      One character of: a, b, c, d
[^ab-d]	      One character except: a, b, c, d
()            Items within parenthesis are retrieved
(a(bc))       Items within the sub-parenthesis are retrieved

REPETITIONS

[ab]{2}       Exactly 2 continuous occurrences of a or b
[ab]{2,5}     2 to 5 continuous occurrences of a or b
[ab]{2,}      2 or more continuous occurrences of a or b
+             One or more
*             Zero or more
?             0 or 1

```python
def extracttext(dataframe,column):
    
    extractedtext=[]
    
    for line in dataframe[column]:
        
        start = line.rfind("/") #get the index of the slash (beginning at the right)
        
        end = line.rfind(".csv") #get the index of ".csv" beginning from the right
        
        extractedstring = line[start+1:end] #extract text between the start and end indexes
        
        extractedtext.append(extractedstring) #append this to the list
    
    return extractedtext #return the full list

#apply the rule

extracttext(files,"filepath")
```

### PRACTICE USING TRANSFORM - when calculating Z scores, we want to return the same number of cells
- so we use TRANSFORM, when we apply the function
- transform allows us to refer to the entire column in our aggregation

```python
# we will define a function that calculates the Z-score of a column

def zscore(column):
    return ((column - column.mean())/column.std())

# apply the function, grouping by origin - to the mpg column

zscore_mpg = auto.groupby("origin")[["mpg"]].transform(zscore)

zscore_mpg.head(10)
```

```python
len(zscore_mpg)
len(auto)

#we can add this as a column to our dataset

auto["zscores"] = zscore_mpg
```

```python
auto.head()
```

```python
# Below we create a function that combines the column transformed with a z-score, with the year and origin columns

def zscore_with_year_and_name(group):
    df=pd.DataFrame(
    {"mpg":zscore(group["mpg"]),
    "year":group["yr"],
    "name":group["name"]})

    return df

resultsset = zscore_with_year_and_name(auto)
resultsset.head()
```

## IMPUTING MISSING VALUES WITH A FUNCTION

```python
titanic = pd.read_csv("Datasets/Course 9 Manipulating Data Frames/titanic.csv")
titanic.head()
```

```python
titanic.info() # note the dataset has 1309 rows, but only 1046 rows have age - we will impute age with median!!
```

```python
#impute a missing column with the median

titanic["ageimpute"] = titanic["age"].fillna(titanic["age"].median())
```

```python
# TO do this as a function

# save the rows with missing values

titanic_missingage = titanic.loc[titanic["age"].isna(),:]
print(len(titanic_missingage)) # 263 rows missing age

def impute_median(dataframe, column):
    dataframe["imputed"]=dataframe[column].fillna(dataframe[column].median())
    return dataframe

# Save the full titanic dataset with imputed age column as "titanic Age_imputed"
titanic_ageimputed = impute_median(titanic,"age")

# Reindex this dataframe using the index from the MINI DATAFRAME ABOVE with rows missing values on age 
#titanic missing age
# Now, we can see those with NANS for "age" now have imputed median age of 28 in the "imputed" column

titanic_ageimputed.reindex(titanic_missingage.index)
```

```python
titanic.head()
```

```python
# the alternative is to fill in with the median of the SUBGROUP I am in

# there are subgroups based on - SEX and CLASS in this dataset

#e.g., 

titanic["ageimpute2"] = titanic["age"].fillna(titanic.groupby(["sex","pclass"])["age"].transform(np.median))

titanic.tail() # Look at CASE 1305 - the imputed MEDIAN differs

# Here we imputed the GROUP median (based on their Sex and Class group)

```

```python
# We can do the above with a function

# Simply create a function that imputes missing values in a series with the median

# but group by SEX and CLASS, and use transform

def fill_median(series):
    result = series.fillna(series.median())
    return result

# below, we create a new column, that imputes the median, grouped by SEX/CLASS combination
# the transform function is operating on the "age" column (a series) --> the age column is the value 
# entered into the "series" part of the function!

titanic["ageimpute3"] = titanic.groupby(["sex","pclass"])["age"].transform(fill_median)

titanic.tail() #note, the last 2 columns get the same result
```

```python
# the other way to do the above is to create a more complex function

# I want to impute missing age values based on "sex and pclass"

def impute_via_median(df,groupingcols,missingcol):
    df["ageimpute4"]=df[missingcol].fillna(df.groupby(groupingcols)[missingcol].transform(np.median))
    return df

impute_via_median(titanic,["sex","pclass"],"age").head()
```

```python
# ALternatively - Make the function simpler
# THe function only imputes missing medians - NOT via groupby
# Use GROUPBY when you want to apply the function

def impute_medians(series):
    return series.fillna(series.median())

# Now apply the function via groupby and transform

titanic["ageimputed5"] = titanic.groupby(["sex","pclass"])["age"].transform(impute_medians)
titanic.head()
```
